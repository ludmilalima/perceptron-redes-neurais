```python
from urllib.request import urlretrieve
import pandas as pd

url_iris = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
urlretrieve(url_iris, 'iris.csv')

df = pd.read_csv(url_iris)
df.columns=['x1','x2','x3','x4','Species']

dfISe = df.assign(Output=0)
dfISe.loc[dfISe.Species == 'Iris-setosa', 'Output'] = 1;
dfISe.loc[dfISe.Species == 'Iris-versicolor', 'Output'] = 0;
dfISe.loc[dfISe.Species == 'Iris-virginica', 'Output'] = 0;

dfIVe = df.assign(Output=0)
dfIVe.loc[dfIVe.Species == 'Iris-setosa', 'Output'] = 0;
dfIVe.loc[dfIVe.Species == 'Iris-versicolor', 'Output'] = 1;
dfIVe.loc[dfIVe.Species == 'Iris-virginica', 'Output'] = 0;

dfIVi = df.assign(Output=0)
dfIVi.loc[dfIVi.Species == 'Iris-setosa', 'Output'] = 0;
dfIVi.loc[dfIVi.Species == 'Iris-versicolor', 'Output'] = 0;
dfIVi.loc[dfIVi.Species == 'Iris-virginica', 'Output'] = 1;

eta = 0.001

class Neuronio:
  def __init__(self, dataset):
    self.x1 = dataset['x1']
    self.x2 = dataset['x2']
    self.x3 = dataset['x3']
    self.x4 = dataset['x4']
    self.w0 = 0
    self.w1 = 0
    self.w2 = 0
    self.w3 = 0
    self.w4 = 0
    self.out = dataset['Output']
    self.iteracoes = 0

'''
i -> índice do registro no dataframe
n -> neuronio que contem os pesos e o dataset proprio
'''
def CalculaResultado(i,n):
  arg = n.w0 + n.x1[i]*n.w1 + n.x2[i]*n.w2 + n.x3[i]*n.w3 + n.x4[i]*n.w4
#  print('\nresultado obtido:', arg, '\nresultado esperado:',n.out[i])
  if arg <= 0:
    return 0
  elif arg > 0:
    return 1

'''
n -> neuronio que contem os pesos a serem recalculados
r -> resultado anterior
'''
def RecalculaPesos(i,n,r):
  n.w0 = n.w0 + (eta*(n.out[i]-r))
  n.w1 = n.w1 + (eta*n.x1[i]*(n.out[i]-r))
  n.w2 = n.w2 + (eta*n.x2[i]*(n.out[i]-r))
  n.w3 = n.w3 + (eta*n.x3[i]*(n.out[i]-r))
  n.w4 = n.w4 + (eta*n.x4[i]*(n.out[i]-r))

nISe = Neuronio(dfISe)
nIVe = Neuronio(dfIVe)
nIVi = Neuronio(dfIVi)

melhorISe = Neuronio(dfISe)
melhorIVe = Neuronio(dfIVe)
melhorIVi = Neuronio(dfIVi)

maxAcertos = 0
maxIndiv = [0,0,0]
acertos = [0,0,0]
iteracoes = 0
treinar = 0

while (treinar <= 500000):
  if iteracoes > 148:
    iteracoes = 0
  if nISe.iteracoes < 150:
    nISe.iteracoes += 1
    resultado = CalculaResultado(iteracoes,nISe)
    if resultado != nISe.out[iteracoes]:
      RecalculaPesos(iteracoes,nISe,resultado)
      acertos[0] = 0
      nISe.iteracoes = 0
    else:
      acertos[0] += 1

  if nIVe.iteracoes < 150:
    nIVe.iteracoes += 1
    resultado = CalculaResultado(iteracoes,nIVe)
    if resultado != nIVe.out[iteracoes]:
      RecalculaPesos(iteracoes,nIVe,resultado)
      acertos[1] = 0
      nIVe.iteracoes = 0
    else:
      acertos[1] += 1

  if nIVi.iteracoes < 150:
    nIVi.iteracoes += 1
    resultado = CalculaResultado(iteracoes,nIVi)
    if resultado != nIVi.out[iteracoes]:
      RecalculaPesos(iteracoes,nIVi,resultado)
      acertos[2] = 0
      nIVi.iteracoes = 0
    else:
      acertos[2] += 1

  iteracoes += 1 
  treinar += 1

  if(acertos[0] + acertos[1] + acertos[2] >= maxAcertos):
    melhorISe = nISe
    melhorIVe = nIVe
    melhorIVi = nIVi
    maxAcertos = acertos[0] + acertos[1] + acertos[2]
    maxIndiv[0] = acertos[0]
    maxIndiv[1] = acertos[1]
    maxIndiv[2] = acertos[2]

print('Máxima combinação de acertos individuais:\nNeurônio 1: ', maxIndiv[0], '\nNeurônio 2: ',maxIndiv[1], '\nNeurônio 3: ',maxIndiv[2],)
print('\nNeuronio 1:\nW0:',melhorISe.w0,'\nW1:',melhorISe.w1,'\nW2:',melhorISe.w2,'\nW3:',melhorISe.w3,'\nW4:',melhorISe.w4)
print('\nNeuronio 2:\nW0:',melhorIVe.w0,'\nW1:',melhorIVe.w1,'\nW2:',melhorIVe.w2,'\nW3:',melhorIVe.w3,'\nW4:',melhorIVe.w4)
print('\nNeuronio 3:\nW0:',melhorIVi.w0,'\nW1:',melhorIVi.w1,'\nW2:',melhorIVi.w2,'\nW3:',melhorIVi.w3,'\nW4:',melhorIVi.w4)
```
